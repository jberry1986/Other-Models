What is MPI, and what other data could we bring in?

MPI is defined at https://en.wikipedia.org/wiki/Multidimensional_Poverty_Index

The global Multidimensional Poverty Index (MPI) is an international measure of acute poverty covering over 100 developing countries. It complements traditional income-based poverty measures by capturing the severe deprivations that each person faces at the same time with respect to education, health and living standards. The MPI assesses poverty at the individual level. If someone is deprived in a third or more of ten (weighted) indicators, the global index identifies them as ‘MPI poor’, and the extent – or intensity – of their poverty is measured by the number of deprivations they are experiencing. The MPI can be used to create a comprehensive picture of people living in poverty, and permits comparisons both across countries, regions and the world and within countries by ethnic group, urban/rural location, as well as other key household and community characteristics.

We can bring in MPI data per region, but does it exist at any level lower? Eg male/female, city etc

Looking at the website; not at a lower level. However there are many more features we could bring in on the region from the MPI sub-regional data file. Lets bring that in.

Also, we need to distinguish between what the loan info telling us about the poverty level, and what the external data can tell us about poverty levels. These are two distinct and different things.

```python
mpi_region_level = tmp_df.groupby(['mpi_region']).agg({'id' : 'count'
                               ,'funded_amount' :  'mean'
                               ,'term_in_months' : 'mean'
                               ,'lender_count' : 'mean'
                               ,'partner_id' : 'nunique'
                               ,'MPI' : 'mean'}).sort_values('id',ascending = False)
mpi_region_level.head()
```

most loans go to the phillipnes

### Are Kiva's loans going to the appropriate places?

```python
#just out of curiosity, what is the correlation between number of lenders to a subnational region, 
    #and what the mpi of that subnational region is

x = mpi_region_level[mpi_region_level['id'] >= 10]['funded_amount']
y = mpi_region_level[mpi_region_level['id'] >= 10]['MPI']

fig, ax = plt.subplots()
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)

from scipy.stats import pearsonr

pearsonr(x,y)
```

seems like there is slight negative correlation between sub national location mpi and the average value of loan sent to that location, although not statistically significant

## What type of poverty exists in a location?

download data from http://ophi.org.uk/multidimensional-poverty-index/global-mpi-2016/

This includes additional information on the types of poverty in each sub-national region, and also the national mpi index

```python
deprivations = pd.read_excel('../data/Tables_5_SubNational_Decomposition_MPI-2017.xlsx'
              ,sheet_name = 2
              ,header = 0
              ,skiprows = 10
              ,skip_footer = 1)

multidimensional = pd.read_excel('../data/Tables_5_SubNational_Decomposition_MPI-2017.xlsx'
                                ,sheet_name = 0
                                ,header = 0
                                ,skiprows = 9
                                ,skip_footer = 5)
```

```python
#drop all NAs
deprivations = deprivations.dropna(axis = 1,how = 'all').copy()
multidimensional = multidimensional.dropna(axis = 1,how = 'all').copy()
```

```python
deprivations.columns= ['ISO country code'
                       ,'Country','Sub-national region'
                       ,'World region','MPI data source','Year','Population Share of the Region'
                       ,'Multidimensional Poverty Index (MPI) of the region'
                       ,'Education','Health','Living standards'
                       ,'Schooling'
                       ,'Child school attendance'
                       ,'Child Mortality'
                       ,'Nutrition'
                       ,'Electricity'
                       ,'Improvedsanitation'
                       ,'Drinking water'
                       ,'Floor'
                       ,'Cooking fuel'
                       ,'Assetownership'
                       ,'Total number of indicators included'
                       ,'Indicator (s) missing']

multidimensional.columns = ['ISO country code'
                           ,'Country','Sub-national region'
                           ,'World region','Survey','Year'
                           ,'Population Share of the Region'
                           ,'Multidimensional Poverty Index (MPI) of the country'
                           ,'Multidimensional Poverty Index(MPI = H*A)'
                           ,'Headcount ratio: Population in multidimensional poverty(H)'
                           ,'Intensity of deprivation among the poor(A) '
                           ,'Destitutes','Proportion of MPI poor who are destitutes'
                           ,'Inequality among the poor'
                           ,'Population vulnerable to poverty(experiencing intensity between 20–32.9% ) '
                           ,'Population in severe poverty(experiencing intensity higher than  50%)'
                           ,'Year of the surveya'
                           ,'Population 2013a'
                           ,'Population 2014a'
                           ,'Number of MPI poor people Population in 2013b'
                           ,'Total number of indicators included'
                           ,'Indicator (s) missing']
```

```python
deprivations.head()
```

```python
#merge the two datasets together
locations = pd.merge(multidimensional,deprivations
         ,on = ['Sub-national region','ISO country code'])
```

```python
reduced = locations.iloc[:,[1,2,3,28,29,30]]
reduced.head().T
```

## Run PCA to see what kind of output we get on our sub national regions

```python
X = np.array(np.array(reduced[['Education','Health','Living standards']]))
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)

from sklearn.decomposition import PCA as sklearnPCA
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X_std)
```

```python
sklearn_pca.explained_variance_ratio_
```

```python
sklearn_pca.components_
```

```python
pd.concat((reduced
           ,pd.DataFrame(X_std,columns = ['a','b','c'])
          ,pd.DataFrame(Y_sklearn)),axis = 1).describe()
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(12, 8))
    plt.scatter(Y_sklearn[:, 0]
        ,Y_sklearn[:, 1])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    #plt.legend(loc='lower center')
    #plt.tight_layout()
    plt.show()
```

can see three distinct groups at least. Worth doing some elbow analysis??

## cluster the output from PCA

```python
from sklearn.cluster import KMeans, SpectralClustering, AgglomerativeClustering

subnational = AgglomerativeClustering(n_clusters=4).fit(Y_sklearn)
a = subnational.labels_

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    plt.scatter(Y_sklearn[:, 0]
        ,Y_sklearn[:, 1]
        ,label = a
        ,c=a
        ,cmap = 'YlOrRd')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    #plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()
```

This scatter plot begins to give us an idea of the different types of locations that there are.

```python
## ADD IN THE PCA ATTRIBUTES

locations['cluster'] = a
locations['PCA1'] = Y_sklearn[:, 0]
locations['PCA2'] = Y_sklearn[:, 1]
```
