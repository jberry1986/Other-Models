```python
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, Imputer
from sklearn.model_selection import train_test_split
import pandas as pd

from sklearn.datasets import load_boston
boston = load_boston()
```

## play around with boston dataset

```python
X = boston['data']
y = boston['target']

scaler = StandardScaler()

X = scaler.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 42)

xgtrain = xgb.DMatrix(X_train,y_train)
xgtest = xgb.DMatrix(X_test,y_test)

param = { 'max_depth':6
         , 'eta':0.3
         , 'subsample' : 0.8
         , 'silent':1
         , 'objective':'reg:linear'
         , 'eval_metric' : 'rmse'}

watchlist = [(xgtest, 'eval'), (xgtrain, 'train')]
num_round = 1000

lrl = [0.2]*1000

bst = xgb.train(param
                , xgtrain
                , num_round
                , watchlist
                , early_stopping_rounds = 2
                , callbacks=[xgb.callback.reset_learning_rate(lrl)])
```

```python
bst.save_model('100_rounds_withSubsample')

bst2 = xgb.Booster(model_file='100_rounds_withSubsample')
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test,bst2.predict(xgtest)))
```
